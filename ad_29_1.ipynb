```python
import scipy.stats as stats
from scipy.stats import pearsonr

# Given summary statistics
mean1 = 10
std1 = 2
n1 = 100

mean2 = 10.5
std2 = 2
n2 = 100


t_stat, p_val = stats.ttest_ind_from_stats(mean1, std1, n1, mean2, std2, n2)


print(f"t-statistic: {t_stat:.4f}")
print(f"p-value: {p_val:.4f}")


alpha = 0.05


if p_val < alpha:
    print("Reject the null hypothesis: The two sample means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample means.")


if t_stat < 0:
    print("Sample 1 likely has a lower mean than Sample 2.")
else:
    print("Sample 1 likely has a higher or equal mean compared to Sample 2.")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(42)


data = np.random.rand(100, 3)
df = pd.DataFrame(data, columns=['A', 'B', 'C'])


sns.set(style="whitegrid")  # Optional for cleaner plot aesthetics
pairplot = sns.pairplot(df, corner=True, diag_kind="kde", markers="o", plot_kws={'alpha': 0.7})


plt.suptitle("Pairplot of Random Data", y=1.02)

plt.show()
```
