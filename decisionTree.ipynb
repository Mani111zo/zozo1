```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as dataset
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn import preprocessing
```

```python
file_name = '/content/iris.csv'
```

```python
iris = pd.read_csv(file_name)
```

```python
iris
```

```python
iris  = iris.rename(columns={'Sepal.Length':'SepalLength','Sepal.Width':'SepalWidth','Petal.Length':'PetalLength','Petal.Width':'PetalWidth'})
```

```python
iris
```

```python
label_encoder = preprocessing.LabelEncoder()
iris['Flower_type'] = label_encoder.fit_transform(iris['Species'])
iris['Flower_type']
```

```python
label_encoder = preprocessing.LabelEncoder()
iris['Species'] = label_encoder.fit_transform(iris['Species'])
iris['Species']
```

```python
x=iris.iloc[:,0:5]
y=iris['Species']
x
```

```python
y=iris.iloc[:,5]
y
```

```python
x=iris.iloc[:,0:4]
y=iris.iloc[:,4]
```

```python
x
```

```python
y
```

```python
columns = list(iris.columns)
columns
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

**Building decision tree classfier (CART) USING Gini criteria

```python
model =DecisionTreeClassifier(criterion='entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
tree.plot_tree(model)
```

```python
fn = ['sepal_length (cm)','sepal_width (cm)','petal_length','petal_width (cm)', 'Species']
cn = ['setosa','versicolor','virginica']
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(model,
               feature_names=fn,
               class_names=cn,
               filled=True)
```

```python
pred =model.predict(x_test)
pd.Series(pred).value_counts()
```

```python
pred
```

```python
pd.crosstab(y_test,pred)
```
